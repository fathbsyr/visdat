## **Import Library**

```python
import matplotlib.pyplot as plt
```

## **Data Set**


```python
list_bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Aug', 'Sept', 'Okt', 'Nov', 'Des']
data_jual_a = [56, 78.90, 43, 45, 67, 12, 48, 80, 12, 24, 50, 90]
data_jual_b = [66, 71.35, 36, 48, 70, 24, 34, 90, 98, 1, 30, 90]
```

## **Single Line Plot**




```python
plt.plot(list_bulan, data_jual_a, label="Produk A")
plt.title('Penjualan Produk A Perbuan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid('true')
plt.show()
```

**Multiple Line Plot and Customizations**

```python
plt.plot(list_bulan, data_jual_a, label="Produk A", linestyle="--", marker="o", color="#00A19C")
plt.plot(list_bulan, data_jual_b, label="Produk B", linestyle="-", marker="x", color="silver")

plt.title('Penjualan Produk A Perbuan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

## **Jenis Garis Untuk menunjukkan Tren**

```python
data_jual_c = [56, 24, 67, 45, 24, 10, 90, 78, 56, 34, 23, 45]
data_jual_d = [67, 34, 78, 23, 45, 66, 77, 12, 34, 56, 89, 99]

plt.plot(list_bulan, data_jual_a, label="Produk A", linestyle="--")
plt.plot(list_bulan, data_jual_b, label="Produk B", linestyle="-")
plt.plot(list_bulan, data_jual_c, label="Produk c", linestyle="dotted")
plt.plot(list_bulan, data_jual_d, label="Produk d", linestyle="-.")


plt.title('Penjualan Produk A Perbuan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid('true')
plt.show()
```

## **Subplot Pada Line Chart**

```python
fig, axs = plt.subplots(2, 1, figsize=(10,8))

axs[0].plot(list_bulan, data_jual_c, label="Produk C", linestyle="--")
axs[0].set_title('Produk Penjualan Produk C Perbulan')
axs[0].set_xlabel('Bulan')
axs[0].set_ylabel('Penjualan')
axs[0].legend()
axs[0].grid('true')

axs[1].plot(list_bulan, data_jual_d, label="Produk D", linestyle="--")
axs[1].set_title('Produk Penjualan Produk D Perbulan')
axs[1].set_xlabel('Bulan')
axs[1].set_ylabel('Penjualan')
axs[1].legend()
axs[1].grid('true')

plt.tight_layout()
plt.show()
```
