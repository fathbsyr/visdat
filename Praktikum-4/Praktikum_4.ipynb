**Multiple Bar Chart**

```python
import matplotlib.pyplot as plt

#data
jurusan = [
    "Ilmu Komputer",
    "Teknik Industri",
    "Sastra Mesin",
    "Teknik Ukir",
    "ADILI JOKOWI"
]

mahasiswa = [120, 180, 200, 89, 1200]
mahasiswa_2024 = [220, 250, 280, 279, 1500]

x = range(len(jurusan))
width = 0.4

plt.bar(x, mahasiswa, width, label="2023", color=["Orange"])
plt.bar([p + width for p in x ], mahasiswa_2024, width=width,
        label="2024", color="skyblue" )

plt.xlabel("Jurusan")
plt.ylabel("Jumlah Mahasiswa")
plt.title("Jumlah Mahasiswa Per Jurusan (2023 vs 2024)")
plt.xticks(rotation=50)
plt.xticks([p + width/2 for p in x], jurusan)
plt.legend()
plt.show()
```

**Bar Chart Untuk Melihat Pola Dan Tren**


```python
import matplotlib.pyplot as plt

#data
jurusan = [
    'Ilmu Komputer',
    'Teknik Industri',
    'Teknik Pertanian',
    'Kehutanan',
    'ADILI JOKOWI'
]

tahun = ["2019", "2020", "2021", "2022", "2023"]

data = {
     'Ilmu Komputer' : [120, 130, 134, 121, 220],
     'Teknik Industri' : [110, 220, 140, 156, 224],
     'Teknik Pertanian' : [124, 179, 156, 178, 156],
     'Kehutanan' : [123, 123, 125, 146, 111],
     'ADILI JOKOWI' : [500, 560, 700, 800, 790]
}

width = 0.16
plt.figure(figsize=(10, 6))
x = range(len(tahun))

plt.bar(x, data['Ilmu Komputer'], width=width, label='Ilmu Komputer', color="blue")
plt.bar([p + width for p in x],
        data['Teknik Industri'], width=width, label='Teknik Indsutri', color="grey")
plt.bar([p + width * 2 for p  in x],
        data['Teknik Pertanian'], width=width, label='Teknik Pertanian', color="green")
plt.bar([p + width * 3 for p in x],
        data['Kehutanan'], width=width, label='kehutanan', color="chocolate")
plt.bar([p + width * 4 for p in x],
        data['ADILI JOKOWI'], width=width, label='ADILI JOKOWI', color="red")

plt.title("Jumlah Mahasiswa Per Jurusan (2019 - 2023)")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Mahasiswa")
plt.xticks(rotation=50)
plt.xticks([p + width * 1.5 for p in x], tahun)
plt.legend()
plt.show()
```
