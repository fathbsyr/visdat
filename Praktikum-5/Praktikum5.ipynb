**Multiple Scatter Plot**

````python
import matplotlib.pyplot as plt

climate = [20, 22, 24, 26, 28, 30, 32, 34, 36]
jualan = [50, 70, 90, 100, 130, 150, 155, 160, 180]
jualan_weekend = [60, 73, 94, 105, 136, 157, 158, 169, 180]

plt.scatter(climate, jualan, color='green', s=80, label='Hari Kerja')
plt.scatter(climate, jualan_weekend, color='blue', s=80, label='Hari Kerja')
plt.title('Penjualan Es Krim Berdasarkan Hari')
plt.xlabel('Suhu (°C)')
plt.ylabel('Penjualan Eskrim')
plt.legend()
plt.show()
````

**Scatter Plot Untuk Melihat Hubungan Data**

````python
import matplotlib.pyplot as plt

climate = [20, 22, 24, 26, 28, 30, 32, 34, 36]
jualan_weekend = [60, 73, 94, 105, 136, 157, 158, 169, 180]

lembap = [60, 65, 70, 75, 80, 85, 90, 95, 100]
titik = [260, 273, 294, 305, 336, 357, 358, 369, 380]

plt.scatter(climate, jualan, c=lembap, s=titik, cmap='coolwarm', alpha=0.6)
plt.title('Hubungan Penjualan, Suhu, Kelembapan')
plt.xlabel('Suhu (°C)')
plt.ylabel('Penjualan Eskrim')
plt.colorbar(label='Kelembapan (%)')
plt.show()
````

**3D Scatterplot**

````python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

data = {
    'Suhu': [20, 22, 24, 26, 28, 30, 32, 34, 36],
    'Penjualan_Weekdays': [50, 60, 70, 80, 90, 100, 110, 120, 130],
    'Penjualan_Weekends': [60, 70, 80, 100, 110, 120, 140, 160, 180]
}
df = pd.DataFrame(data)
suhu = df['Suhu']
penjualan_weekdays = df['Penjualan_Weekdays']
penjualan_weekends = df['Penjualan_Weekends']
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(
    suhu,
    penjualan_weekdays,
    penjualan_weekends,
    color='orange',
    s=80,
    alpha=0.7,
    label='Data Penjualan'
)
ax.set_title('Hubungan Penjualan Es Krim dengan Suhu (3D ver)')
ax.set_xlabel('Suhu')
ax.set_ylabel('Penjualan Weekdays')
ax.set_zlabel('Penjualan Weekends')
plt.tight_layout()
ax.legend()
plt.show()
````
