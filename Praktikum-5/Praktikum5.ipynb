**Scatter Plot**

```python
import matplotlib.pyplot as plt

climate = [20, 22, 24, 26, 28, 30, 32, 34, 36]
jualan = [50, 70, 90, 100, 130, 150, 155, 160, 180]

plt.scatter(climate, jualan, color='red')
plt.title('Grafik Korelasi')
plt.xlabel('Suhu (°C)')
plt.ylabel('Penjualan Eskrim')
plt.show()
```

```python
import matplotlib.pyplot as plt

climate = [20, 22, 24, 26, 28, 30, 32, 34, 36]
jualan = [50, 70, 90, 100, 130, 150, 155, 160, 180]

plt.scatter(climate, jualan, color='blue', s=100, edgecolor='black', alpha=0.7)
plt.title('Grafik Korelasi')
plt.xlabel('Suhu (°C)')
plt.ylabel('Penjualan Eskrim')
plt.grid(True)
plt.show()
```

**Menambahkan Scatter Plot**

```python
import matplotlib.pyplot as plt

climate = [20, 22, 24, 26, 28, 30, 32, 34, 36]
jualan = [50, 70, 90, 100, 130, 150, 155, 160, 180]
jualan_weekend = [60, 73, 94, 105, 136, 157, 158, 169, 180]

plt.scatter(climate, jualan, color='green', s=80, label='Hari Kerja')
plt.scatter(climate, jualan_weekend, color='blue', s=80, label='Hari Kerja')
plt.title('Penjualan Es Krim Berdasarkan Hari')
plt.xlabel('Suhu (°C)')
plt.ylabel('Penjualan Eskrim')
plt.legend()
plt.show()
```

**Scatter Plot Untuk Melihat Hubungan Data**

```python
import matplotlib.pyplot as plt

climate = [20, 22, 24, 26, 28, 30, 32, 34, 36]
jualan_weekend = [60, 73, 94, 105, 136, 157, 158, 169, 180]

lembap = [60, 65, 70, 75, 80, 85, 90, 95, 100]
titik = [260, 273, 294, 305, 336, 357, 358, 369, 380]

plt.scatter(climate, jualan, c=lembap, s=titik, cmap='coolwarm', alpha=0.6)
plt.title('Hubungan Penjualan, Suhu, Kelembapan')
plt.xlabel('Suhu (°C)')
plt.ylabel('Penjualan Eskrim')
plt.colorbar(label='Kelembapan (%)')
plt.show()
```
