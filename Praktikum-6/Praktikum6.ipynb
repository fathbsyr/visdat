```python
import matplotlib.pyplot as plt
import numpy as np
```

**Dataset**

```python
stores = ['Store A', 'Store B', 'Store C']
laki = [150, 200, 180]
cewek = [120, 230, 170]

q1_laki = [150, 180, 160]
q1_cewek = [140, 200, 180]
q2_laki = [170, 190, 175]
q2_cewek = [150, 250, 120]
q3_laki = [172, 192, 177]
q3_cewek = [152, 252, 122]
q4_laki = [173, 193, 178]
q4_cewek = [153, 253, 123]
```

**Basic Stacked**

```python
x = np.arange(len(stores))
plt.bar(x, laki, label='Pria', color='grey')
plt.bar(x, cewek, bottom=laki, label='Wanita', color='pink')
plt.xlabel('Toko')
plt.title('Populasi Antara Gender dan Toko')
plt.ylabel('Populasi')
plt.xticks(x, stores)
plt.legend()
```

**Kustomisasi Stacked Bar Chart**

```python
x = np.arange(len(stores))
plt.bar(x, laki, label='Pria', color='grey')
plt.bar(x, cewek, bottom=laki, label='Wanita', color='pink')
plt.xlabel('Toko')
plt.title('Populasi Antara Gender dan Toko')
plt.ylabel('Populasi')
plt.xticks(x, stores)
plt.legend()

for i in range(len(x)):
    plt.text(x[i], laki[i]/2, str(laki[i]), ha='center', color='white')
    plt.text(x[i], laki[i] + cewek[i]/2, str(cewek[i]), ha='center', color='black')
```

**Multiple Stacked Bar Chart**

```python
width = 0.2
x = np.arange(len(stores))
plt.bar(x - width * 1.5, q1_laki, label='Q1 Pria', color='lightblue', width=width)
plt.bar(x - width * 1.5, q1_cewek, bottom=q1_laki, label='Q1 Wanita', color='pink', width=width)

plt.bar(x - width/2, q2_laki, label='Q2 Pria', color='blue', width=width)
plt.bar(x - width/2, q2_cewek, bottom=q2_laki, label='Q2 Wanita', color='red', width=width)

plt.bar(x + width/2, q3_laki, label='Q3 Pria', color='green', width=width)
plt.bar(x + width/2, q3_cewek, bottom=q3_laki, label='Q3 Wanita', color='yellow', width=width)

plt.bar(x + width * 1.5, q4_laki, label='Q4 Pria', color='purple', width=width)
plt.bar(x + width * 1.5, q4_cewek, bottom=q4_laki, label='Q4 Wanita', color='orange', width=width)

plt.title('Populasi Antara Gender dan Toko (multiple)')
plt.xlabel('Toko')
plt.ylabel('Sales')
plt.xticks(x, stores)
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()
```
